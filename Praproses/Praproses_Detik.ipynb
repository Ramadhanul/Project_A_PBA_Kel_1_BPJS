```python
import re

# Hapus kalimat "ADVERTISEMENTSCROLL TO CONTINUE WITH CONTENT" di kolom teks
df['cleaned_text'] = df['text'].apply(lambda x: re.sub(r"ADVERTISEMENTSCROLL TO CONTINUE WITH CONTENT", "", str(x)))

# Tampilkan hasil pembersihan
df[['text', 'cleaned_text']].head()
```

```python
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from collections import Counter

# Download stopwords dan tokenizer NLTK (hanya perlu dijalankan sekali)
nltk.download('punkt')
nltk.download('stopwords')

# 1. Mengonversi teks ke huruf kecil
df['cleaned_text'] = df['cleaned_text'].str.lower()

# 2. Frekuensi kata sebelum Tokenisasi, Penghapusan Stopwords, dan Stemming
all_words_before = ' '.join(df['cleaned_text']).split()
word_freq_before = Counter(all_words_before)

# Menampilkan frekuensi kata sebelum preprocessing
print("Frekuensi Kata Sebelum Preprocessing:")
print(word_freq_before.most_common(10))  # Menampilkan 10 kata teratas

# Visualisasi frekuensi kata sebelum preprocessing
plt.figure(figsize=(12, 6))
plt.bar(*zip(*word_freq_before.most_common(10)))
plt.title('Frekuensi Kata Sebelum Preprocessing')
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Tokenisasi

from nltk.tokenize import word_tokenize

# Tokenisasi teks yang sudah dibersihkan
df['tokens'] = df['cleaned_text'].apply(word_tokenize)

# Tampilkan hasil tokenisasi
print(df[['cleaned_text', 'tokens']].head())
```

```python
# 2. Penghapusan Stopwords

# Inisialisasi stopwords bahasa Indonesia
stop_words = set(stopwords.words("indonesian"))

# Hapus stopwords dari kolom tokens
df['tokens_without_stopwords'] = df['tokens'].apply(
    lambda x: [word for word in x if word not in stop_words and word.isalpha()]
)

# Tampilkan hasil penghapusan stopwords
print(df[['tokens', 'tokens_without_stopwords']].head())
```

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming
def stemming(tokens):
    return ' '.join([stemmer.stem(word) for word in tokens])

# Terapkan stemming pada kolom tokens tanpa stopwords
df['processed_text'] = df['tokens_without_stopwords'].apply(stemming)

# Tampilkan contoh sebelum dan sesudah stemming
for i in [0]:  # Ganti 0 dengan index yang Anda inginkan
    original_text = ' '.join(df['tokens_without_stopwords'].iloc[i])
    stemmed_text = df['processed_text'].iloc[i]
    print(f"Original: {original_text}")
    print(f"Stemmed:  {stemmed_text}")
    print()
```

```python
# 4. Frekuensi kata setelah Tokenisasi, Penghapusan Stopwords, dan Stemming
all_words_after = ' '.join(df['processed_text']).split()
word_freq_after = Counter(all_words_after)

# Menampilkan frekuensi kata setelah preprocessing
print("\nFrekuensi Kata Setelah Preprocessing:")
print(word_freq_after.most_common(10))  # Menampilkan 10 kata teratas

# Visualisasi frekuensi kata setelah preprocessing
plt.figure(figsize=(12, 6))
plt.bar(*zip(*word_freq_after.most_common(10)))
plt.title('Frekuensi Kata Setelah Preprocessing')
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengunduh data yang sudah diproses ke file CSV dengan pemisah ;
output_file = 'detik_processed_bpjs_articles.csv'
df.to_csv(output_file, sep=';', index=False)

# Untuk mengunduh di Google Colab
from google.colab import files
files.download(output_file)
```
