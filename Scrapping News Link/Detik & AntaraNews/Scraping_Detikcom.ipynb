```python
from bs4 import BeautifulSoup
import requests
import time

# Step 1: Fungsi untuk mengekstrak data dari setiap halaman
def extract_data_from_page(page_num):
    url = f"https://www.detik.com/search/searchnews?query=BPJS&page={page_num}&result_type=relevansi&siteid=55"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to retrieve page {page_num}")
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    articles = soup.find_all('article', class_="list-content__item")

    data = []
    for article in articles:
        # Judul
        title_tag = article.find('h3', class_='media__title')
        title = title_tag.get_text(strip=True) if title_tag else "No title"

        # Hanya ambil artikel yang mengandung kata "BPJS" dalam judul
        if "BPJS" not in title:
            continue

        # Link
        link = title_tag.find('a')['href'] if title_tag and title_tag.find('a') else "No link"

        # Kategori
        category_tag = article.find('h2', class_='media__subtitle')
        category = category_tag.get_text(strip=True) if category_tag else "No category"

        # Tanggal
        date_tag = article.find('div', class_='media__date')
        date = date_tag.get_text(strip=True) if date_tag else "No date"

        data.append({
            'title': title,
            'link': link,
            'category': category,
            'date': date
        })

    return data

# Step 2: Looping melalui semua halaman
all_data = []
total_pages = 81  # Menentukan jumlah halaman yang ingin ditelusuri

for page in range(1, total_pages + 1):
    print(f"Extracting data from page {page}")
    page_data = extract_data_from_page(page)
    all_data.extend(page_data)

    # Penundaan untuk menghindari permintaan terlalu cepat
    time.sleep(2)

# Step 3: Output hasil
for item in all_data:
    print(f"Title: {item['title']}")
    print(f"Link: {item['link']}")
    print(f"Category: {item['category']}")
    print(f"Date: {item['date']}")
    print("-" * 50)
```

```python
import pandas as pd
# Step 1: Mengonversi hasil ke dalam DataFrame
df = pd.DataFrame(all_data)

# Step 2: Mengekspor DataFrame ke CSV dengan pemisah ;
df.to_csv('detik_bpjs_articles.csv', sep=';', index=False)

print("Data berhasil diekspor ke 'detik_bpjs_articles.csv' dengan pemisah ;")
```
