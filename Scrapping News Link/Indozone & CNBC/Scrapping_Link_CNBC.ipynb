```python
import requests
from bs4 import BeautifulSoup

def scrape_website(base_url, max_pages):
    scraped_data = []

    for page in range(1, max_pages + 1):
        # Buat URL untuk setiap halaman
        page_url = f"{base_url}&page={page}"
        response = requests.get(page_url)
        soup = BeautifulSoup(response.content, "html.parser")

        # Ambil semua artikel dari halaman tersebut
        articles = soup.find_all('article')

        for article in articles:
            # Ambil judul artikel
            title_element = article.find('h2', class_='font-semibold text-23 group-hover:text-cnbc-primary-blue')
            if title_element:
                title = title_element.text.strip()

                # Cari artikel terkait BPJS
                if 'bpjs' in title.lower():
                    # Ambil link artikel
                    link_element = article.find('a')
                    if link_element:
                        link = link_element.get('href', '')

                        # Ambil tanggal artikel
                        date_element = article.find('span', class_='text-xs text-gray')
                        if date_element:
                            date = date_element.text.strip()

                            # Tambahkan ke list jika link belum ada
                            if not any(d['Link'] == link for d in scraped_data):
                                scraped_data.append({
                                    'Title': title,
                                    'Link': link,
                                    'Date': date
                                })

    return scraped_data

# Contoh penggunaan
base_url = "https://www.cnbcindonesia.com/search?query=bpjs"
max_pages = 388
data = scrape_website(base_url, max_pages)

# Print hasilnya
for item in data:
    print(item)
```

```python
def save_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
```

```python
save_to_csv(data, 'link_cnbc.csv')
```
